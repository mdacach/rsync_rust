```python
import pandas as pd

full_data = pd.read_csv("linux_chunk_size_to_compression_ratio.csv")
```

```python
full_data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='chunk_size', y='compression_ratio', palette='flare', hue='compression_ratio', data=full_data)
plt.xlabel('block size used (in bytes)')
plt.ylabel('compression rate')
plt.title("block size x compression rate")
```
